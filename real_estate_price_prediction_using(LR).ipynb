```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/real_estate_price_size.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
x=data[['size']]
y=data[['price']]
plt.scatter(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_predict=lr.predict(x_test)
```

```python
print("Intercept:",lr.intercept_)
print("Coefficient:",lr.coef_)
```

```python
from sklearn.metrics import r2_score
r_squared=r2_score(y_test,y_predict)
print("R2_SCORE:",r_squared)
```

```python
lr.predict([[1275.46],[854]])
```

```python
plt.scatter(x_train,y_train,color='blue')
plt.plot(x_test,y_predict,color='red')
```

```python
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import root_mean_squared_error as rmse
print("mean_absolute_error:",mae(y_test,y_predict))
print("mean_squared_error:",mse(y_test,y_predict))
print("root_mean_squared_error:",rmse(y_test,y_predict))
```
